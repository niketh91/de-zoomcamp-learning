```python
#Some data fields are also of type object instead of datetime. So need to fix data types

dtype = {
    "VendorID": "Int64",
    "passenger_count": "Int64",
    "trip_distance": "float64",
    "RatecodeID": "Int64",
    "store_and_fwd_flag": "string",
    "PULocationID": "Int64",
    "DOLocationID": "Int64",
    "payment_type": "Int64",
    "fare_amount": "float64",
    "extra": "float64",
    "mta_tax": "float64",
    "tip_amount": "float64",
    "tolls_amount": "float64",
    "improvement_surcharge": "float64",
    "total_amount": "float64",
    "congestion_surcharge": "float64"
}

parse_dates = [
    "tpep_pickup_datetime",
    "tpep_dropoff_datetime"
]

df = pd.read_csv(
    url,
    dtype=dtype,
    parse_dates=parse_dates
)
```

```python
df.dtypes
```

```python
# we add SQL Alchemy from notebook instead of terminal by adding a ! before the uv command

!uv add sqlalchemy psycopg2-binary
```

```python
# Create data base connection

from sqlalchemy import create_engine
engine = create_engine('postgresql://root:root@localhost:5432/ny_taxi')
```

```python
conn = engine.connect()
print("Connected!")
conn.close()
```

```python
# DDL schema

print(pd.io.sql.get_schema(df, name='yellow_taxi_data', con=engine))
```

```python
#Let's just create the table without any data in it. For it use df.head(0)
# head(n=0) makes sure we only create the table, we don't add any data yet
df.head(n=0).to_sql(name='yellow_taxi_data', con=engine, if_exists='replace')
```

```python
len(df)
```

```python
# Since data size is large, lets split in into chunks and ingest data chunk by chunk

df_iter = pd.read_csv(
    url,
    dtype=dtype,
    parse_dates=parse_dates,
    iterator = True,
    chunksize = 100000
)
```

```python
#Iterate over chunks
for df_chunk in df_iter:
    print(len(df_chunk))
```

```python
# to see progress, lets add tqdm

!uv add tqdm
```

```python
# ingest/insert in chunks
from tqdm.auto import tqdm

for df_chunk in tqdm(df_iter):
     df_chunk.to_sql(name='yellow_taxi_data', con=engine, if_exists='append')
```

```python
# if above is not running, do it without tqdm

for df_chunk in df_iter:
     df_chunk.to_sql(name='yellow_taxi_data', con=engine, if_exists='append')
     print(f'Inserted: {len(df_chunk)}')
```
